```python
import ast

#function to extract first three cast members from each movie
def counter(obj):
    L = []
    count=0
    for i in obj:
        if count!=3:
            L.append(obj[count])
            count+=1
        else:
            break
    return L
     
```

```python
import ast
#tmdb['cast']=tmdb['cast'].apply(ast.literal_eval)
tmdb['cast'] = [ast.literal_eval(x) for x in tmdb['cast']]
```

```python
tmdb['keywords'] = [ast.literal_eval(x) for x in tmdb['keywords']]
```

```python
tmdb['director'] = [ast.literal_eval(x) for x in tmdb['director']]
```

```python
tmdb['genres'] = [ast.literal_eval(x) for x in tmdb['genres']]
```

```python
tmdb['cast']=tmdb['cast'].apply(counter)
```

```python
tmdb.drop(['vote_average'],axis=1)
```

```python
#function to remove spaces from the names of cast and crew

def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
tmdb['keywords']=tmdb['keywords'].apply(collapse)
```

```python
tmdb['genres']=tmdb['genres'].apply(collapse)
```

```python
tmdb['cast']=tmdb['cast'].apply(collapse)
```

```python
tmdb['director']=tmdb['director'].apply(collapse)
```

```python
tmdb['overview']=tmdb['overview'].astype('object')
tmdb.dropna(inplace=True)
```

```python
tmdb['overview'].dropna(inplace=True)
tmdb['overview'] = tmdb['overview'].apply(lambda x:x.split())
```

```python
tmdb['overview']
```

```python
tmdb['tags'] = tmdb['overview'] + tmdb['genres'] + tmdb['keywords'] + tmdb['cast'] + tmdb['director']
tmdb_new=tmdb.drop(['overview','genres','keywords','cast','director','original_language','vote_average'],axis=1)
```

```python
tmdb_new['tags'].astype('string')
```

```python
tmdb['cast'][10]
```

```python
tmdb_new['tags']=tmdb_new['tags'].apply(lambda x:" ".join(x))
```

```python
tmdb_new['tags']=tmdb_new['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vectors = cv.fit_transform(tmdb_new['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(vectors)
```

```python
similarity_score
```

```python
from fuzzywuzzy import fuzz

def recommend(movie):
    closest_match = max(tmdb_new['title'], key=lambda x: fuzz.ratio(x.lower(), movie.lower()))
    movie_index = tmdb_new.index[tmdb_new['title'] == closest_match].tolist()[0]
    #movie_index = tmdb_new[tmdb_new['title']==movie].index[0]
    distances = similarity_score[movie_index]
    movies_list = sorted((enumerate(distances)),reverse=True,key=lambda x:x[1])[1:10]
    top_n = [(tmdb_new.iloc[i]['title'], movies_list[i][1]) for i in range(1, 5+1)]

    for m in movies_list:
        print(tmdb_new.iloc[m[0]].title,closest_match)
    
```

```python
recommend('avengers')
```
